## Adding in bbox prediction

```python
model = torch.hub.load("ultralytics/yolov5", "custom", "trained_models/2_classes.pt")
```

```python
test_image = cv2.imread("test_image/57502_000480_Endzone_frame0495.jpg")
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
result = model(test_image)
```

```python
result.xyxy[0].numpy()[0]
```

```python
def draw_bboxes(frame, bboxes:list):
    color_map = {
        0:(255,0,0),
        1:(0,255,0)
    }
    for bbox in bboxes:
        cv2.rectangle(frame, (round(bbox[0]),round(bbox[1])), (round(bbox[2]),round(bbox[3])), color_map[bbox[-1]], 2)
        
    return frame
```

```python
video_reader = cv2.VideoCapture("test_video/57583_000082_Endzone.mp4")
codex = cv2.VideoWriter_fourcc(*"MP4V")
fps = 30
frame_size=  (1280, 720)
video_writer = cv2.VideoWriter("test_video/with_bboxes.mp4", codex, fps, frame_size)
try:
    while True:
        retVal, frame = video_reader.read()
        if retVal is None:
            break
        
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pred = model(frame)
        frame_w_bboxes = draw_bboxes(frame, pred.xyxy[0].numpy())
        frame_w_bboxes = cv2.cvtColor(frame_w_bboxes, cv2.COLOR_RGB2BGR)
        video_writer.write(frame_w_bboxes)
        
#         cv2.imshow("player", frame_w_bboxes)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
except Exception as e:
    print(e)
finally:
    video_writer.release()
    video_reader.release()
    cv2.destroyAllWindows()
```
